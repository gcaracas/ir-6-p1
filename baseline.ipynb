```python
stop_words = set(stopwords.words('english')) 
```

```python
def remove_stop_words(array=[]):
    global stop_words
    token_dict = {}
    count=0;
    total=[]
    tokenizer=TweetTokenizer(strip_handles=False, reduce_len=False, preserve_case=False)
    for w in array:
        if pd.isnull(w): continue
        w_t = w.translate(string.punctuation)
        t=tokenizer.tokenize(w_t)
        word=[]
        for sp in t:
            if not sp in stop_words:
                word.append(sp.lower())
        if len(word) > 0: 
            total.append(word)
            token_dict[count]=total
            count=count+1
    return token_dict
```

```python
q1=remove_stop_words(array=q1)
q2=remove_stop_words(array=q2)
```
